```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
b=0.25
n0=1
t=np.linspace(0,10,101)
```

```python
def function(n,t):
    dndt=b*n
    return dndt
```

```python
n=odeint(function,n0,t)
```

```python
plt.plot(t,n)
plt.title('my plot')
plt.xlabel('time')
plt.ylabel('y')
plt.show()
```

```python
n1=10
b=float(2/8) 
d=float(1/2)
m=float(3/10)
t1=np.linspace(0,10,101)
```

```python
def ode(n,t):
    dndt = b*n - d*n + m
    return dndt
```

```python
n=odeint(ode,n1,t1)
```

```python
plt.plot(t1,n)
plt.title('my_plot')
plt.ylabel('y')
plt.xlabel('time')
plt.show()
```

```python
y0=[95.0,5.0]
a=0.025
c=1.0
t=np.linspace(0,10,101)
```

```python
def derivate(y,t):
    s,n = y
    dydt1 = -a*c*n*s
    dydt2 = a*c*s*n
    return [dydt1,dydt2]
```

```python
sol = odeint(derivate,y0,t)
```

```python
plt.plot(t,sol[:,0],'b',label="healthy")
plt.plot(t,sol[:,1],'r',label='infected')
plt.legend(loc='best')
plt.xlabel('time(day)')
plt.ylabel('y')
plt.show()
```
