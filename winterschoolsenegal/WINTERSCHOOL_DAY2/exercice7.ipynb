```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
r = 0.01
y0=0.8
t=np.arange(0,250)
```

```python
def expgro(y,t):
    n = y
    dydt = r*n
    return dydt
```

```python
sol = odeint(expgro,y0,t)
```

```python
plt.plot (t, sol [: ,0] , linewidth =3)
plt.xlabel ('Time (years)',fontsize =20)
plt.ylabel ('Population', fontsize =20)
plt.title ('Solution of exponential model', fontsize =24)
plt.show()
plt.savefig ('ExpGrowth.pdf')
```
